```python
import pandas as pd
import numpy as np
```

```python
reviews = pd.read_excel('tripadvisor.xlsx', 'reviews', index=False, na_values=['NA'])
reviews = reviews.drop(["id","date",'title','text','helpfulness','total_points'],axis=1)
reviews.head()
```

```python
selected = reviews[reviews['type'] == 'Hotels'].index.tolist()
reviews = reviews.drop(selected)
```

```python
reviews['taObjectCity'].value_counts()
```

```python
london = reviews[reviews['taObjectCity'] == 'London']
barcelona = reviews[reviews['taObjectCity'] == 'Barcelona']
madrid = reviews[reviews['taObjectCity'] == 'Madrid']
hamburg = reviews[reviews['taObjectCity'] == 'Hamburg']
names = ["London","Barcelona","Madrid","Hamburg"]

rowCount = [len(london), len(barcelona),len(madrid),len(hamburg)]
userCount = [len(london.username.unique()),len(barcelona.username.unique()),len(madrid.username.unique()),len(hamburg.username.unique())]
itemCount = [len(london.taObject.unique()),len(barcelona.taObject.unique()),len(madrid.taObject.unique()),len(hamburg.taObject.unique())]
d = {'Cities':pd.Series(names), 'TotalRatings':rowCount, 'Users':pd.Series(userCount), 'Objects':itemCount}
df = pd.DataFrame(d)
df
```
